```python
import pandas as pd
import numpy as np
import os
from baseline import run_pipeline
# 假设你的 PROCESSED_DIR 和 PRED_DIR 已经定义好了
PROCESSED_DIR = "../processed_data"
PRED_DIR = "../predictions"
os.makedirs(PROCESSED_DIR, exist_ok=True)
os.makedirs(PRED_DIR, exist_ok=True)

# 1. 加载 numpy 数据
X_train_raw = np.load("../datasets/psm/PSM_train.npy")
X_test_raw  = np.load("../datasets/psm/PSM_test.npy")
y_test_raw  = np.load("../datasets/psm/PSM_test_label.npy")

# 2. 定义列名
# 因为列名是基于长度的数字范围，我们直接生成 0 到 N-1 的列表
feature_cols = list(range(X_train_raw.shape[1]))
target_col = 'label'

# 3. 构建 DataFrame
train_full_df = pd.DataFrame(X_train_raw, columns=feature_cols)
test_df = pd.DataFrame(X_test_raw, columns=feature_cols)

# 4. 处理标签 (Labels)
# 训练集全为正常点，填充 0
train_full_df[target_col] = 0
train_df = train_full_df.copy()  # 训练集已经包含标签，直接使用
# 测试集拼上真实的 y_test，使用 ravel() 防止 shape 为 (N, 1) 导致的格式警告
test_df[target_col] = y_test_raw.ravel()

# 5. 划分验证集 (Validation Set)
# PSM 是时间序列数据，建议不要打乱顺序 (shuffle)，直接取最后 20% 作为验证集
val_df = test_df

# 6. 运行你的 Pipeline
results_df = run_pipeline(
    dataset_name="PSM",
    train_df=train_df,
    val_df=val_df,
    test_df=test_df,
    feature_cols=feature_cols,
    target_col=target_col
)

print(results_df)
```
